```python
from model import Actor, Critic
from torchviz import make_dot
import torch
RANDOM_SEED = 2
STATE_SIZE = 24
ACTION_SIZE = 2
```

```python
actor = Actor(state_size=STATE_SIZE, action_size=ACTION_SIZE, seed=RANDOM_SEED, fc1_units=128, fc2_units=128)
critic = Critic(state_size=STATE_SIZE, action_size=ACTION_SIZE, seed=RANDOM_SEED, fcs1_units=128, fc2_units=128)
```

```python
x = torch.randn(1, STATE_SIZE)
y = actor(x)

make_dot(y.mean(), params=dict(actor.named_parameters()))
```

```python
state = torch.randn(1, STATE_SIZE)
action = torch.randn(1, ACTION_SIZE)
y = critic(state, action)

make_dot(y.mean(), params=dict(critic.named_parameters()))
```
